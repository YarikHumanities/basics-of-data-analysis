1. Виділити один зі стовпців (на вибір) з файлу як об’єкт Series, 

```python
import numpy as np
import pandas as pd
our_field ='easting'
series = pd.Series(pd.read_csv('frogs.csv', skipinitialspace=True, usecols=[our_field])[our_field])
print(series)
```

Виділити з
нього підмасив

```python
series = series[4:25]
print(series)
```

Задати назви індексів цього об’єкту.

```python
index_values = ['raw_' + str(x) for x in range(series.size)]
print(index_values)
series.index=index_values
print(series)
```

Виділити підмасиви
за допомогою прямої та непрямої індексацій.

```python
loc_custom_indexing_series = series.loc['raw_10':'raw_15']
print(loc_custom_indexing_series)
iloc_indexing_series = series.iloc[10:14]
print(iloc_indexing_series)
```

```python
size_and_weight = pd.DataFrame({"id": [0, 1, 2, 3, 4, 6],
                             "size": [10, 15, 20, 12, 13, 25],
                             "weight": [100, 200, 300, 400, 500, 600]})
size_and_weight.set_index("id")
size_and_weight

sex = pd.DataFrame({"sex": ["male", "female", "female", "male", "male", "male"], "id": [0, 1, 2, 3, 4, 7]})
sex.set_index("id")

population_1 = pd.DataFrame({"quantity": [1000, 2000, 3000, 4000, 5000, 6000, 7000]}, index=["first", "second", "third", "fourth", "five", "six", "seven"])
population_2 = pd.DataFrame({"quantity": [10000, 7000]}, index=["tenth", "seven"])

sample_1 = pd.DataFrame([["Kyiv", 1000], ["Odessa", 2000], ["Kharkiv", 3000], ["Uzhhorod", 4000]], columns=["City", "population"])
sample_2 = pd.DataFrame([["Lviv", 5000], ["Lutsk", 6000], ["Vynnytsa", 7000]], columns=["City", "population"])
```
